## Random number

```python
%matplotlib inline
# hist a set of random numbers, Gaussian distributed
import numpy as np
import matplotlib.pyplot as plt
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

N = 10000000
x = np.random.random(N)
ax1.hist(x, 50, normed=True, color='#A60628', alpha=0.6);

x = np.random.randn(N)
ax2.hist(x , 50, normed=True, color='#348ABD', alpha=0.6);

plt.suptitle("Histogram of random numbers drawn from uniform and Gaussian distribution",fontsize=20);
```

## Random walk

```python
N  = 10000; # number of steps
rw = np.cumsum(1-2*np.random.randint(0, 2, N))     # radom walk code
plt.xlabel('N'); plt.ylabel('R');plt.plot(rw); 
```

```python
N = 400        # step length
Nw = 10**6     # number of random walks
R = np.zeros(Nw)
for i in range(Nw):
    x = np.cumsum(1-2*np.random.randint(0, 2, N))
    R[i] = x[-1]/(np.sqrt(N))
```

```python
plt.hist(R, bins=50, normed=True, histtype='step', color='#A60628', linewidth=2, label='R')

xx = np.random.randn(10000)
plt.hist(xx, bins=20, normed=True, alpha=0.4, label='Gaussian')
plt.legend()
plt.show()
```
